## Working with Scikit-learn neural network classifier architectures (Multi-layer Perceptron) for predicting broken pipes

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.neural_network import MLPClassifier
from crossval import crossval_by_year
```

```python
df = pd.read_csv('final_data.csv') 
```

```python
params = {
    'learning_rate_init': 0.001,
    'alpha': 0,
    'random_state': 42,
    'hidden_layer_sizes': (75, 50, 25),
    'verbose': True,
    'max_iter': 50
}

avg_acc = crossval_by_year(df, MLPClassifier, params, normalize_data=True)
```

```python
params = {
    'learning_rate_init': 0.001,
    'alpha': 0,
    'random_state': 42,
    'hidden_layer_sizes': (100, 75, 25),
    'verbose': True,
    'max_iter': 50
}

avg_acc = crossval_by_year(df, MLPClassifier, params, normalize_data=True)
```
